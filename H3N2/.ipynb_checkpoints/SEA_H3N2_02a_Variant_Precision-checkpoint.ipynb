# Import libraries

```python
from os.path import expanduser
from importlib.machinery import SourceFileLoader
from matplotlib.lines import Line2D 

import matplotlib.pyplot as plt
import pandas as pd 
import re
import os
import itertools
import subprocess

# load common flu and ngs libraries 
laeb_lib = expanduser("../PYTHONLIB/")
fc = SourceFileLoader('fc', "%s/flu_common.py"%(laeb_lib)).load_module()
ngs = SourceFileLoader('ngs', "%s/laeb_ngs_pipeline.py"%(laeb_lib)).load_module()
```

# Inputs 

```python
# inputs 
# file path to data folder - fastq files to be analysed must be in {data folder}/raw
data_folder = './data' 
# reference fasta file name (should be placed in input_folder)
ref_fasta_fname = './input/H3N2_Bris07.fasta' 
# CSV file containing the CDR regions of each gene segment (numbering should be based on that of the given reference sequence)
cds_coords = "./input/CDS_H3N2_Bris07.csv"
# primer coordinates (sequence numbering must be based on given reference sequence)
primer_coords = "./input/H3N2_primer_coords.csv"
# file name for metadata file. 
meta_fname = './results/metadata_w_covbreadth.csv' 

# variant calling 
min_cov = 100 # minimum coverage 
min_var_prop = 0.02 # minimum variant proportion 
min_var_freq = 0 # minimum number of variants to be called 
err_tol = 0.01 # threshold to which variant called could result from base calling error 
```

# Read metadata and variant calling files

```python
# metadata 
## metadata must have 'sampid' header which is used as sample identifier 
meta_df = pd.read_csv('%s'%meta_fname)
sorted_sampid = sorted(set(meta_df['sampid']))

meta_df = meta_df.set_index("sampid")
display (meta_df.head())

# vcf 
variant_call_df = pd.read_csv("./results/variant_call_MinCoV%i_MinProp%.2f_MinFreq%i_ErrTol%.2f.csv"%(min_cov, min_var_prop, min_var_freq, err_tol))
variant_call_df = variant_call_df.set_index(["sampid", 'gene', "nucpos"])
display (variant_call_df.head())
```

# Tally base counts by amplicon

For each intrahost single nucleotide variant (iSNV) in every sample, we tally base counts attributed to overlapping amplicons.

```python
if not os.path.isdir("./results/freqacc"): 
    os.mkdir("./results/freqacc")

for sampid in sorted_sampid: 
    
    if os.path.isfile("./results/freqacc/freqacc_%s.csv"%(sampid)): 
        continue 
        
    # skip if there are no iSNVs found in sample
    try: 
        sample_vcf = variant_call_df.loc[sampid]
    except: 
        continue 
    
    precision_read_df= {}
    # read mapped reads dataframe 
    mapped_reads_df = pd.read_csv("./results/mapped/mapped_%s.csv"%(sampid))
    mapped_reads_df = mapped_reads_df[~mapped_reads_df['gene'].isin(["7-M", "8-NS"])]
    
    # get all genes where variants were called 
    var_gene_list = list(set(sample_vcf.index.get_level_values(0)))
    
    # for each gene where variants were called 
    for gene in var_gene_list:
        # filter dataframes 
        gene_mapped_reads_df = mapped_reads_df[mapped_reads_df["gene"]==gene]
        gene_sample_vcf = sample_vcf.loc[gene]
        
        # for each called variant position 
        for pos in gene_sample_vcf.index: 
            # filter mapped reads with position 
            pos_gene_mapped_reads_df = gene_mapped_reads_df[(gene_mapped_reads_df["start_nucaln_pos"]<=pos)&(gene_mapped_reads_df["end_nucaln_pos_plus1"]>pos)]
            # filter only reads where amplicon_nr had been identifed
            pos_gene_mapped_reads_df = pos_gene_mapped_reads_df[~pd.isna(pos_gene_mapped_reads_df["amplicon_nr"])]
            if len(pos_gene_mapped_reads_df) == 0: 
                continue 
            # variant must be present in at least two overlapping amplicons
            if len(set(pos_gene_mapped_reads_df["amplicon_nr"])) < 2: 
                continue 
                
            # for each read, tally base counts for each called variant sorted by amplicon 
            for r, read in pos_gene_mapped_reads_df.iterrows():
                amplicon = read.amplicon_nr
                read_base = read.seq[pos-read.start_nucaln_pos]
                if re.search("^(A|T|G|C)$", read_base, re.I):
                    try: 
                        precision_read_df[gene][pos][amplicon][read_base] += 1
                    except: 
                        try: 
                            precision_read_df[gene][pos][amplicon][read_base] = 1
                        except: 
                            try: 
                                precision_read_df[gene][pos][amplicon] = {read_base:1}
                            except: 
                                try: 
                                    precision_read_df[gene][pos] = {amplicon:{read_base:1}}
                                except: 
                                    precision_read_df[gene] = {pos:{amplicon:{read_base:1}}}
    
    # continue if there are no variants found in overlapping amplicons for sample
    if len(precision_read_df) == 0: 
        continue 
    # save to file 
    precision_read_df = pd.DataFrame.from_dict({(i,j,k): precision_read_df[i][j][k]
                                                for i in precision_read_df.keys()
                                                for j in precision_read_df[i].keys()
                                                for k in precision_read_df[i][j].keys()},
                                               orient='index')
    precision_read_df.index.names = ['Gene', 'Position', "Amplicon"]
    precision_read_df = precision_read_df.fillna(0)
    precision_read_df.to_csv("./results/freqacc/freqacc_%s.csv"%(sampid))
```

# Plot precision plots

Here, we plot frequencies of iSNVs found in overlapping amplicons against each other, stratifying them by the viral load (i.e. cycle threshold values) of the samples, to assess the precision of our variant calls.

```python
ct_to_xy = {}

for sampid in sorted_sampid: 
    # for each sample 
    try: 
        precision_read_df = pd.read_csv("./results/freqacc/freqacc_%s.csv"%(sampid))
    except: 
        continue 
    
    # read base tallies stratified by amplicon 
    precision_read_df = precision_read_df.set_index(['Gene', 'Position', "Amplicon"])
    precision_read_df['coverage'] = precision_read_df.sum(axis=1)
    # must satisfy min_cov within each amplicon 
    precision_read_df = precision_read_df[precision_read_df['coverage']>min_cov]
    
    precision_read_df = precision_read_df.reset_index()
    precision_read_df = precision_read_df.set_index(['Gene', 'Position']).sort_index()
    
    # get cycle threshold value 
    ct = meta_df.loc[sampid, "ct"]
    
    # for each iSNV
    for idx in list(set(precision_read_df.index)): 
        idx_precision_read_df = precision_read_df.loc[idx]
        
        # skip if iSNV is only (mostly) covered by one amplicon (i.e. the second amplicon did not satisfy min_cov)
        if len(idx_precision_read_df) < 2: 
            continue 
        
        if isinstance(idx_precision_read_df, pd.Series): 
            continue
        
        # get reported frequencies of iSNV (i.e. combined across all amplicons)
        idx_vcf = variant_call_df.loc[(sampid, idx[0], idx[-1])]
        
        # for each pair of overlapping amplicons
        for amp1, amp2 in itertools.combinations(idx_precision_read_df['Amplicon'], 2): 
            amp1_stats = idx_precision_read_df[idx_precision_read_df["Amplicon"]==amp1]
            amp2_stats = idx_precision_read_df[idx_precision_read_df["Amplicon"]==amp2]
            
            base_list = [base for base in list(amp2_stats) if re.search("^(A|T|G|C)$", base)]
            
            amp1_freq = (amp1_stats[base_list]/amp1_stats["coverage"].iloc[0]).to_numpy()[0]
            amp2_freq = (amp2_stats[base_list]/amp2_stats["coverage"].iloc[0]).to_numpy()[0]
            
            for _f, f1 in enumerate(amp1_freq):
                f2 = amp2_freq[_f]
                
                # frequencies in each amplicon must also satisfy min_var_prop 
                if f1 < min_var_prop and f2 < min_var_prop:
                    continue
                # only analyse minority iSNVs 
                if (f1 > 0.5 and f2 < 0.5) or (f2 > 0.5 and f1 < 0.5):
                    continue 
                # compute variant frequencies for each amplicon 
                base = base_list[_f]
                if idx_vcf['nuc_var'].iloc[0] == base: # variant could be a variant or consensus in the other amplicon 
                    r = idx_vcf['nuc_prop'].iloc[0]
                    
                elif idx_vcf['nuc_consensus'].iloc[0] == base:
                    r = idx_vcf['nuc_consensus_freq'].iloc[0]/idx_vcf['nuc_coverage'].iloc[0]
                
                else: 
                    display (base, idx_vcf)
                    print (idx_vcf['nuc_var'].iloc[0])
                    raise Exception
                
                # save results for plotting 
                try:
                    ct_to_xy[ct]['x'].append(f1)
                    ct_to_xy[ct]['y'].append(f2)
                    ct_to_xy[ct]['r'].append(r)
                except: 
                    ct_to_xy[ct] = {'x':[f1], 'y':[f2], 'r':[r]}
                    
```

```python
color_list = ["#ae76a3", "#882e72", '#1965b0', '#5289c7', 
              '#7bafde', '#4eb265', '#90c987', '#cae0ab', 
              '#f7f056', '#f6c141', '#f1932d', '#e8601c', '#dc050c']

with plt.style.context('default'): 
    legend_elements = []
    fig, axes = plt.subplots(1, 2, figsize=(11.7, 5))
    
    ax = axes[0]
    for idx, ct in enumerate(sorted(ct_to_xy.keys())[::-1]): 
        #print (ct)
        sp = ax.scatter(ct_to_xy[ct]['x'], 
                        ct_to_xy[ct]['y'], 
                        alpha=0.5,
                        color=color_list[idx], label=ct)
        legend_elements.append(sp)
    
    ax.set_xlabel('Amplicon 1 frequency', size=14)
    ax.set_ylabel('Amplicon 2 frequency', size=14)
    
    ########################################################
    ax = axes[1]
    for idx, ct in enumerate(sorted(ct_to_xy.keys())[::-1]): 
        #print (ct)
        ax.scatter(ct_to_xy[ct]['x'], 
                        ct_to_xy[ct]['r'], 
                        alpha=0.5,
                        color=color_list[idx])
        
    for idx, ct in enumerate(sorted(ct_to_xy.keys())[::-1]): 
        #print (ct)
        ax.scatter(ct_to_xy[ct]['y'], 
                        ct_to_xy[ct]['r'], 
                        alpha=0.5,
                        color=color_list[idx])
    
    ax.set_xlabel('Amplicon frequency', size=14)
    ax.set_ylabel('Combined frequency', size=14)
    
    for ax in axes: 
        ax.plot([0, 1], [0, 1], c='k', alpha=0.6, ls='--')
        
        ax.set_xlim(-0.01, 1.01)
        ax.set_ylim(-0.010, 1.01)
        
        ax.tick_params(which='both', labelsize=12)
    
    # legend 
    fig.legend(handles=legend_elements[::-1], 
               loc='center left', bbox_to_anchor=(.9, 0.5), 
               fontsize=12, title='CT values')
    
    plt.savefig('./results/figures/freq_error_plot.png', dpi=600, bbox_inches='tight', pad_inches=0.01)
    plt.show()
```
